# Popularity Based


```python
ratings_with_id = rating.merge(chitwan, on="restaurant-id")
```

```python
ratings_with_id.shape
ratings_with_id.head(0)
```

```python
num_rating_df = ratings_with_id.groupby('Title').count()['Ratings'].reset_index()
num_rating_df.rename(columns={ "Ratings":"num_ratings"}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_id.groupby('Title').mean()['Ratings'].reset_index()
avg_rating_df.rename(columns={ "Ratings":"avg_ratings"}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="Title")
popular_df.sort_values('num_ratings', ascending=False)
```

```python
popular_df = popular_df[popular_df['num_ratings']>=4].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(chitwan, on="Title").drop_duplicates('Title')[['Title', 'location', 'Image_URL','num_ratings' , 'avg_ratings']]
```

# Collaborative Filtering


```python
x = ratings_with_id.groupby('User-ID').count()['Ratings'] >2
critics = x[x].index
```

```python
# x.shape
critics.shape
```

```python
filtered_rating = ratings_with_id[ratings_with_id['User-ID'].isin(critics)] 
```

```python
y = filtered_rating.groupby('Title').count()['Ratings']>=2
famous_restro = y[y].index
```

```python
famous_restro.shape
```

```python
final_ratings = filtered_rating[filtered_rating['Title'].isin(famous_restro)]
```

```python
pt=final_ratings.pivot_table(index="Title", columns="User-ID", values='Ratings')
```

```python
pt.fillna(0,inplace=True)
```

```python
 pt.sort_index(ascending=False)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(restaurant_name):
    index = np.where(pt.index==restaurant_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x:x[1], reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item=[]
        temp_df = chitwan[chitwan['Title']== pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Title')['Title'].values))
        item.extend(list(temp_df.drop_duplicates('Title')['location'].values))
        item.extend(list(temp_df.drop_duplicates('Title')['Image_URL'].values))
        
        data.append(item)
        
    return data    
```

```python
recommend('Hungry Eye Restaurant')
```
